```python
#Import dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#read in csv dta
df = pd.read_csv('myopia.csv')
df
```

```python
#Drop Myopic column
df.drop(columns=['MYOPIC'], inplace=True)
df
```

```python
#Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
X_scaled.shape
```

```python
#Apply Dimensionality Reduction
pca = PCA(n_components=.9)
pca.fit(X_scaled)
principal_components = pca.transform(X_scaled)
principal_components.shape
```

```python
#preserve 90% of the explained variance in dimensionality reduction
pca.explained_variance_ratio_.sum()
```

```python
#Further reduce the dataset dimensions with t-SNE and visually inspect the results
tsne = TSNE(perplexity=50)
features = tsne.fit_transform(principal_components)
features.shape
```

```python
#Create a scatter plot of the t-SNE output
plt.scatter(features[:,0], features[:,1])
```

### There appears to be 4 disctinct clusters from the scatter plot results

```python
#Use a for loop to determine the inertia for each k between 1 through 10.
inertia = list()
k = list(range(1,9))

for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(principal_components)
    inertia.append(km.inertia_)
    
elbow_df = pd.DataFrame({'k': k, 'inertia': inertia})
```

```python
#Create an elbow plot to identify the best number of clusters
elbow_df.plot.line(x='k', y='inertia')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
```

### According to the above elbow graph, the patients can be clustered into 3 to 4 separate groups.
